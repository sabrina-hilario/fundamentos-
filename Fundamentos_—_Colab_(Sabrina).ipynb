 O que é um notebook?

 Através do notebook é possível a estruturação de um projeto , ele age como uma pasta que mantém ligado todos os blocos de códigos,ele facilita a separação dos códigos de diferentes projetos.




 O que é uma célula?

A célula é uma unidade menor de organização dos códigos e textos,é dentro da célula que os códigos e textos serão trabalhados.Durante a execução de uma célula, ela executará ao mesmo tempo todos os códigos que estão dentro dela,por isso é recomendado a criação de várias células, como forma de identificar com mais rapidez a função dela.
