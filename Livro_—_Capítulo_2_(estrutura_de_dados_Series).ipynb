Resumo

As estruturas de dados são métodos de armazenamento e de manipulação de informação, a ED Series é composta por um vetor de dados e por um vetor de rótulos, o vetor de dados é responsável pelas características do objeto, enquanto o vetor rótulos é o próprio nome ou categoria daquele objeto . As Series podem ser criadas por dicionários ou por listas, na lista o índice (indicador da posição do elemento) precisa ser inteiro, mas no dicionário não necessariamente, também não existe conceito de ordem nos dicionários, uma vez que eles são coleções não ordenadas de pares chave:valor. Assim, as Series vão unir a sequência ordenada de elementos da lista com o acesso rápido pelos rótulos dos dicionários. Os dtype definem o ´´tipo de coisa `` que está guardado dentro da Series , o vetor dado de uma Series possui sempre o mesmo dtype. A indexação é uma forma de obter ou acessar o elemento de uma Series, existem três tipos de indexação.




Exemplos do capítulo

```python


import pandas as pd

notas = pd.Series([7.6, 5.0, 8.5, 9.5, 6.4])
lst_matriculas = ['M02','M05','M13','M14','M19']
lst_nomes = ['Bob','Dayse','Bill','Cris','Jimi']
alunos = pd.Series(lst_nomes,index=lst_matriculas)
print(notas); print("---------------"); print(alunos)
```

```python
import pandas as pd
alunos = pd.Series({'M02':'Bob','M05':'Dayse','M13':'Bill',
'M14':'Cris','M19':'Jimi'})
alunos.name = "alunos"
alunos.index.name = "matrículas"
print(alunos)
print('-----------')
tamanho = alunos.size
dados = alunos.values
rotulos = alunos.index
alunos_tipo = type(alunos)
alunos_dtype = alunos.dtype
alunos_idx_dtype = alunos.index.dtype
print('número de elementos: ', tamanho)
print('vetor de dados: ', dados)
print('vetor de rótulos: ', rotulos)
print('tipo (type): ', alunos_tipo)
print('dtype da Series:', alunos_dtype)
print('dtype do vetor de rótulos:', alunos_idx_dtype)
```

```python
import pandas as pd
import numpy as np
s1 = pd.Series([2,4,6])
s2 = pd.Series([1,3,5])
print('s1:'); print(s1)
print('s2:'); print(s2)
print('---------------------------')
print('s1 * 2')
print(s1 * 2)
print('---------------------------')
print('s1 + s2')
print(s1 + s2)
print('---------------------------')
print('raiz quadrada dos elementos de s1')
print(np.sqrt(s1))
```

5) Crie um programa que armazene nomes e salários de funcionários e exiba apenas os funcionários com salário maior ou igual a 3000, usando indexação booleana.

Salários = (2500,1800,3200,4100,2900)

Funcionários = ( F01:Ana,F02:Carlos,F03:Fernanda,F04:Marcos,F05:Juliana )

```python
import pandas as pd
salarios = pd.Series([2500,1800,3200,4100,2900])
funcionarios = pd.Series({'F01':'Ana','F02':'Carlos','F03':'Fernanda','F04':'Marcos','F05':'Juliana'})
idx_aprovados = salarios[salarios >= 3000].index
print('relação de funcionários com salários maiores ou iguais a 3000:')
print('---------------------------')
print(funcionarios[idx_aprovados])
```
